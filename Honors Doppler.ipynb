# The Relativistic Doppler Effect

## Introduction

Many of us have experienced the familiar phenomenon of a sound's pitch changing as its source moves toward or away from us - like the changing pitch of a passing ambulance siren. This phenomenon, known as the Doppler effect, occurs when there is relative motion between a wave source and an observer.

While the classical Doppler effect explains frequency shifts for sound waves in a medium like air, the situation becomes more complex when we consider light waves. Unlike sound, light doesn't require a medium to propagate, and its speed remains constant for all observers according to Einstein's special relativity. These fundamental differences lead to the relativistic Doppler effect, which incorporates both the relative motion between source and observer and the relativistic effects of time dilation.

## Comparing Classical and Relativistic Doppler Effects

### Classical Doppler Effect

For sound waves traveling in a medium, the frequency shift depends on the separate velocities of both the source and observer relative to the medium. If a sound wave propagates at velocity $v$ through a medium, with source and observer moving at velocities $v_s$ and $v_o$ relative to that medium, the observed frequency $f'$ relates to the emitted frequency $f$ by:

$$f' = f \cdot \frac{v + v_o}{v - v_s}$$

This formula works well for sound waves but breaks down when applied to light, which has no reference medium.

### Relativistic Doppler Effect

For light waves, we must incorporate the Einstein's postulates :
- The speed of light is constant for all observers
- There is no preferred reference frame
- Time dilates for objects in relative motion

These principles lead to a fundamentally different formulation of the Doppler effect.

## Derivation from Special Relativity Principles

Consider a light source moving away from an observer with velocity $v$. In the source's rest frame, it emits waves with period $T_0 = 1/f_0$.

Due to the source's motion during this emission, two effects combine:

1. The spatial effect: As the source moves away, successive wave crests are emitted from increasingly distant positions
2. The temporal effect: Time dilation causes the observer to perceive the source's clock as running slower

For a source emitting $n$ waves during a time interval $T$ in the observer's frame, the total length of the wave train is:

$$\text{Length of wave train} = cT \mp vT$$

Where we use minus for approaching and plus for receding.

Since there are $n$ waves in this train, the wavelength is:

$$\lambda = \frac{cT \mp vT}{n}$$

The observed frequency is $f = c/\lambda$, giving:

$$f = \frac{cn}{cT \mp vT}$$

In the source's rest frame, the number of waves emitted is:

$$n = f_0 T_0$$

Where $T_0$ is the proper time in the source frame. Due to time dilation, this relates to the observer's time by:

$$T_0 = \frac{T}{\gamma}$$

Substituting these relations and simplifying leads to our relativistic Doppler formula for a source moving toward or away from an observer along their line of sight:

$$f = f_0 \cdot \frac{1}{\gamma(1 \mp \beta)}$$

Where:
- $f$ is the frequency measured by the observer
- $f_0$ is the frequency in the source's rest frame
- $\beta = v/c$ is the relative speed as a fraction of light speed
- $\gamma = 1/\sqrt{1-\beta^2}$ is the Lorentz factor
- **The minus sign applies when the source and observer approach each** other
- **The plus sign applies when they recede from each other**

This can be rewritten in a more elegant form:

$$f = f_0 \sqrt{\frac{1 \pm \beta}{1 \mp \beta}}$$

Where the upper signs apply for approaching and lower signs for receding.

## General Case (Arbitrary Angle)

For the general case where the motion occurs at an arbitrary angle $\theta$ to the line of sight, the frequency shift is:

$$f = f_0 \frac{1-\beta\cos\theta}{\sqrt{1-\beta^2}}$$

## Special Cases

### Longitudinal Doppler Effect

When the relative motion is directly along the line connecting source and observer, we have:

For approaching source and observer:
$$f = f_0 \sqrt{\frac{1+\beta}{1-\beta}}$$

For receding source and observer:
$$f = f_0 \sqrt{\frac{1-\beta}{1+\beta}}$$

## Significance of the Transverse Doppler Effect

The transverse Doppler effect deserves special attention because it has no classical counterpart and offers a direct demonstration of relativistic time dilation. When a source moves perpendicular to an observer's line of sight, classical physics predicts no frequency shift. However, special relativity predicts that the observed frequency will be lower than the emitted frequency due to time dilation.

$$f = f_0 \gamma^{-1} = f_0\sqrt{1-\beta^2}$$

This frequency shift arises solely from time dilation and serves as direct evidence for special relativity's time dilation effect.


## Visualizing the Relativistic Doppler Effect

Let's create visualizations to better understand how the relativistic Doppler effect behaves under different conditions.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline
'''
these are here just to make the plots readable,
was having problem where the text was going into 
the differnt axes
'''
plt.rcParams['font.size'] = 14
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['axes.titlesize'] = 18

plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14



plt.rcParams['legend.fontsize'] = 14
plt.rcParams['figure.figsize'] = [10, 8]



# rel shift func
def relativistic_doppler_shift(beta, theta):
    gamma = 1 / np.sqrt(1 - beta**2)
    return gamma * (1 - beta * np.cos(theta))

# Plot 1 F shift vs angle
betas = np.linspace(0, 0.99, 500)
angles = [0, np.pi/4, np.pi/2, 3*np.pi/4, np.pi]
angle_labels = ['0 (approaching)','45','90 (transverse)', '135', '180 (receding)']

plt.figure(figsize=(12, 8))
for angle, label in zip(angles, angle_labels):
    shift = relativistic_doppler_shift(betas, angle)
    plt.plot(betas, shift, label=label, linewidth=2)
#make graph nice:
plt.title('Relativistic Doppler Effect: Frequency Shift vs. Velocity')
plt.xlabel(r'Relative velocity ($\beta = v/c$)') #just figured out you can latex graph text, neat
plt.ylabel(r'Frequency ratio ($f/f_0$)')


plt.grid(True, alpha=0.3)
plt.legend()
plt.ylim(0, 5)
plt.tight_layout()
plt.show()
#====================================================
#P2 classic vs rel
beta_compare = np.linspace(0, 0.9, 500)

classical_approaching = 1 / (1 - beta_compare)
relativistic_approaching = np.sqrt((1 + beta_compare)/(1 - beta_compare))

classical_receding = 1 / (1 + beta_compare)
relativistic_receding = np.sqrt((1 - beta_compare)/(1 + beta_compare))


#make graph nice:
plt.figure(figsize=(12, 8))
plt.plot(beta_compare, classical_approaching, 'b-', label='Classical (approaching)', linewidth=2)
plt.plot(beta_compare, relativistic_approaching, 'b--', label='Relativistic (approaching)', linewidth=2)
plt.plot(beta_compare, classical_receding, 'r-', label='Classicsl (receding)', linewidth=2)
plt.plot(beta_compare, relativistic_receding, 'r--', label='Relativistic (receding)', linewidth=2)

plt.title('Classical vs. Relativistic Doppler Effect')
plt.xlabel(r'Relative velocity ($\beta = v/c$)')
plt.ylabel(r'Frequency ratio ($f/f_0$)') 
plt.grid(True, alpha=0.3)
plt.legend()
plt.ylim(0, 5)
plt.tight_layout()
plt.show()
#====================================================
# p3 3D graph for a bit of nuance 
beta_mesh = np.linspace(0, 0.9, 100)
theta_mesh = np.linspace(0, np.pi, 100)


beta_grid, theta_grid = np.meshgrid(beta_mesh, theta_mesh)

shift_grid = relativistic_doppler_shift(beta_grid, theta_grid)
#make graph nice:
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(beta_grid, theta_grid * 180/np.pi, shift_grid, 
                       cmap=cm.viridis, linewidth=0, antialiased=True)
#fix axes before submit
ax.set_title('Relativistic Doppler Effect: 3D Visualization')
ax.set_xlabel(r'Relative velocity ($\beta = v/c$)')
ax.set_ylabel(r'Angle $\theta$ (degrees)')
ax.set_zlabel(r'Frequency ratio ($f/f_0$)')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)
plt.tight_layout()
plt.show()
```

The first plot demonstrates how the frequency shift changes with velocity for different observation angles. As a source approaches an observer (0), the observed frequency increases significantly as the velocity approaches the speed of light. For a receding source (180), the frequency decreases, approaching zero as velocity increases.

The second plot compares classical and relativistic Doppler formulas. At low velocities, the formulas give similar results, but they diverge significantly as velocity increases. The relativistic formula properly accounts for time dilation, which becomes increasingly important at higher velocities.

The third plot provides a 3D visualization showing how the frequency shift depends simultaneously on both velocity and observation angle, giving us a comprehensive view of the relativistic Doppler effect.


## Conclusion

The relativistic Doppler effect stands as one of the most practical and observable consequences of Einstein's special relativity. By incorporating time dilation and the constancy of light speed, it presents a fundamentally different view of frequency shifts compared to the classical understanding.

Key differences between classical and relativistic Doppler effects:

1. The relativistic effect depends only on the relative motion between source and observer, with no reference to an absolute medium
   
2. The relativistic effect includes the transverse Doppler shift, which has no classical equivalent
   
3. At low velocities, the relativistic formulas approximate the classical ones, maintaining correspondence
