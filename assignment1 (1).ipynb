### Questions


Recall that the speed of light in vacuum is 
\begin{equation}
c = \frac{1}{\sqrt{\epsilon_0 \mu_0}}, \; \; \; \; \; (1)
\end{equation}
where $\epsilon_0 = 8.85 \times 10^{-12}$ F/m (Farads per meter)$^{*}$ is the electric permittivity of free space and $\mu_0 = 1.26 \times 10^{-6}$ N/A$^2$ (Newtons per Ampere$^2$)$^{**}$ is the magnetic permeability of free space. The electric field created by a point charge $q$ is
\begin{equation}
E = \frac{1}{4 \pi \epsilon_0} \frac{q}{r^2}. \; \; \; \; (2)
\end{equation}

Suppose a charged particle is moving at some fraction $f$ of the speed of light in vacuum ($v_q = f c; \; 0 < f < 1$) as seen in reference frame $K$. Reference frame $K^{\prime}$ is co-moving with the point charge, so that $\frac{dx^{\prime}}{dt^{\prime}} = v^{\prime}_q = 0$ in frame $K^{\prime}$. To make the math easier, put $q$ at the origin of $K^{\prime}$.

To understand Einstein's motivation for developing the special theory of relativity, make the <strong>incorrect, erroneous, and wrong</strong> assumption that the Galilean velocity transformation holds true for all values of $f$:
\begin{equation}
\frac{dx^{\prime}}{dt^{\prime}} = \frac{dx}{dt} - v_q \; \; \; (3)
\end{equation}

</ol>

```python
import numpy as np
import matplotlib.pyplot as plt

electronCharge = 1.602e-19  # Coulombs
epsilonZero = 8.85e-12     # F/m
    
   
rPrime = np.logspace(-12, -10, 1000)  # 1 picometer to 100 picometers    
electricField = (1/(4*np.pi*epsilonZero))* (electronCharge/(rPrime**2))
    
  
plt.figure(figsize=(10, 6))
plt.plot(rPrime*1e12, electricField*1e-12, 'b-', linewidth=2)  #these factors seemed to be the most graphable
#TeraNewtons per Coulomb and Picometers 

#plt.xscale('log')
#plt.yscale('log')
#I don't know if you would prefer to see this log scaled?
    
plt.xlabel("Distance r' (picometers)")
plt.ylabel("Electric Field (TN/C)") 
plt.title("Electric Field vs Distance from Point Charge\nin Frame K'")
plt.grid(True, which="both", ls="-", alpha=0.2)

    
plt.show()

```

```python
epsilonZero = 8.85e-12  #F/m
electronCharge = 1.602e-19  #C
rPrime = 1e-6  #1m 

fValues = np.linspace(0, 0.99, 1000)

gamma = 1 / np.sqrt(1 - fValues**2)

epsilonPrime = epsilonZero * gamma
electricFieldPrime = (1/(4*np.pi*epsilonPrime)) * (electronCharge/rPrime**2)


fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))
fig.tight_layout(pad=3.0)

# ' vs f
ax1.plot(fValues, epsilonPrime, 'b-', linewidth=2)
ax1.set_xlabel('f (fraction of c)')
ax1.set_ylabel("' (F/m)")
ax1.set_title("Electric Permittivity in K' Frame")
ax1.grid(True, alpha=0.2)

#E' vs f
ax2.plot(fValues, electricFieldPrime, 'r-', linewidth=2)
ax2.set_xlabel('f (fraction of c)')
ax2.set_ylabel("E' (V/m)")
ax2.set_title("Electric Field in K' Frame at r' = 1 m")
ax2.grid(True, alpha=0.2)



plt.show()
```
